# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing and exploring dataset

```python
# Assuming the dataset is saved as "unemployment_in_india.csv" in your working directory
data = pd.read_csv('Unemployment in India.csv')
```

```python
# View the first few rows of the dataset
print(data.head())
```

```python
# Get information about the dataset (e.g., column types, missing values)
print(data.info())
```

```python
# Check for missing values
print(data.isnull().sum())
```

# Cleaning

```python
# Drop rows with missing values
data_cleaned = data.dropna()

# Verify the new number of non-null entries after handling missing values
print(data_cleaned.info())
```

```python
print(data_cleaned.columns)
```

# Visualization

```python
# Remove leading space from the ' Date' column name
data.rename(columns={' Date': 'Date'}, inplace=True)

# Convert the 'Date' column to a date/time data type
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Date', y=data[' Estimated Unemployment Rate (%)'], marker='o')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Trends')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of region-wise average unemployment rate
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Region', y=' Estimated Unemployment Rate (%)')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Region-wise Average Unemployment Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of unemployment rate distribution by frequency
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x=' Frequency', y=' Estimated Unemployment Rate (%)')
plt.xlabel('Frequency')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Distribution by Frequency')
plt.show()
```

```python
# Calculating correlation matrix
corr_matrix = data[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Violin plot of unemployment rate distribution by region
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Region', y=' Estimated Unemployment Rate (%)')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Distribution by Region')
plt.xticks(rotation=45)
plt.show()
```
